# Exmaple of Conditional statement... how to add column and fill the column 

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\umohd\OneDrive\Desktop\Jupyter\Use If Else\Student Marks.csv")
```

```python
data   # import data and want to add result column and  declared who is pass and fail.
```

```python
def fill_pass_or_fail(Marks):
    if Marks>=60:           # if the students scored more than 60 marks, he/she is declared passed.
        return "Pass"
    else:
        return "Fail"
```

```python
fill_pass_or_fail(60)
```

```python
data ["Result"] = data ["Marks"].apply(fill_pass_or_fail)     # adding result column  with function
```

```python
data
```

```python
data.to_csv("Updated_data.csv")   
```

# Exmaple of for loop....  how to fill NaN column and Correct MRP price column 

```python
import pandas as pd
data = pd.read_excel(r"C:\Users\umohd\OneDrive\Desktop\Jupyter\Use If Else\Sales Sheet.xlsx")
```

```python
data    # import data
```

```python
for index in data.index:
    if data.loc[index, "MRP"]< 500:  # if any MRP is less than 500 rupees then it has to be done from 999 rupees
        data.loc[index, "MRP"] = 999
    if pd.isna(data.loc[index, "MRP"]):
        data.loc[index,"MRP"] = 1199   # replace NAN column into 1199 rupees
```

```python
data
```

```python
data.to_csv("Updated_sales_data.csv")
```
